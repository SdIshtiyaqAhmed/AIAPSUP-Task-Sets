**Description:** Few-shot: Provide 2–3 examples to generate a function that formats full names as “Last, First”.

**Expected Output:** Well-structured function respecting the examples

**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Prompt:**
write a function that formats full name as "last, first"

Example1:

Input: Jhon Doe

Output: Doe, Jhon

Example2:

Input: Test User Name

Output: User, Name Test

**Gemini (AI) Generated Code:**

```python
def format_name(full_name):
  """
  Formats a full name as 'last, first'.

  Args:
    full_name: A string representing the full name.

  Returns:
    A string representing the formatted name.
  """
  names = full_name.split()
  if len(names) > 1:
    return names[-1] + ", " + " ".join(names[:-1])
  else:
    return full_name

# Example usage:
print(format_name("Jhon Doe"))
print(format_name("Test User Name"))
```
