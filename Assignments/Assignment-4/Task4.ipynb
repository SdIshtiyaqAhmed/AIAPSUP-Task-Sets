**Description:** Compare zero-shot and few-shot prompts for writing a function that counts the number of vowels in a string.

**Expected Output:** Functional output and comparative reflection

**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Zero-shot Prompting**

**Prompt:** write a function to count the number of vowels in a string

**Gemini (AI) Generated Code:**

```python
def count_vowels(text):
  """
  Counts the number of vowels in a string.

  Args:
  text: The input string.

  Returns:
  The number of vowels in the string.
  """
  vowels = "aeiouAEIOU"
  count = 0
  for char in text:
    if char in vowels:
      count += 1
  return count

# Example usage:
print(count_vowels("Hello World"))
```

**Few-shot Prompting**

**Prompt:** write a function to count the number of vowels in a string.

Example1:

Input: WeLcOmE

Output: 3

Example2: 

Input: Computer Science Engineering (CSE)

Output: 12

Example3:

Input: HTML

Output: 0

```python
def count_vowels(text):
  """
  Counts the number of vowels in a string (case-insensitive).

  Args:
    text: The input string.

  Returns:
    The number of vowels in the string.
  """
  vowels = "aeiouAEIOU"
  count = 0
  for char in text:
    if char in vowels:
      count += 1
  return count

# Example usage:
print(count_vowels("WeLcOmE"))
print(count_vowels("Computer Science Engineering (CSE)"))
print(count_vowels("HTML"))
```

**Observations:** The logic of both the functions count_vowels() remains the same in zero-shot and few-shot prompting, but in the zero-shot prompting there is only a single example usage in the code and in the few-shot prompting there are multiple example usages in the code that are based on the examples given in the prompt.
