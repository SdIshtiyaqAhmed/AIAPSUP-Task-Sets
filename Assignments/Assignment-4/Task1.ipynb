**Description:** Zero-shot: Prompt AI to write a function that checks whether a given year is a leap year.

**Expected Output:** AI-generated function with no examples provided

**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Prompt:** write a function to check whether a given year is a leap year or not

**Gemini (AI) Generated Code:**

```python
def is_leap(year):
  """
  Checks if a given year is a leap year.

  Args:
    year: An integer representing the year.

  Returns:
    True if the year is a leap year, False otherwise.
  """
  if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    return True
  else:
    return False

# Example usage:
print(is_leap(2000))
print(is_leap(1900))
print(is_leap(2024))
```
