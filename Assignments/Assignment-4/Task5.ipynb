**Description:** Use few-shot prompting to generate a function that reads a .txt file and returns the number of lines.

**Expected Output:** Working file-processing function with AI-guided logic

**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Prompt:** write a function that returns the number of lines in a text file.

Example1:

Input: 

sample1.txt:

Hello World!

Output: 1

Example2:

Input:

Sample2.txt:

Welcome
to
Computer Science Engineering (CSE)

Output: 3

Example3:

Input:

Sample3.txt:

Output: 0

**Gemini (AI) Generated Code:**

```python
def count_lines_in_file(filename):
  """
  Counts the number of lines in a text file.

  Args:
    filename: The path to the text file.

  Returns:
    The number of lines in the file.
  """
  try:
    with open(filename, 'r') as f:
      line_count = sum(1 for line in f)
    return line_count
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
    return -1 # Or raise an exception, depending on desired behavior

# Example usage:
print(f"sample1.txt: {count_lines_in_file('sample1.txt')}")
print(f"sample2.txt: {count_lines_in_file('sample2.txt')}")
print(f"sample3.txt: {count_lines_in_file('sample3.txt')}")
print(f"non_existent_file.txt: {count_lines_in_file('non_existent_file.txt')}")
```
