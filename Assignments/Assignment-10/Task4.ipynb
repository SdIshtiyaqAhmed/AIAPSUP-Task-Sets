**Gemini (AI) Generated Code:**

````python
def calculate_total(scores):
    total = 0
    for s in scores:
        total += s
    return total

def calculate_average(scores):
    total = calculate_total(scores)
    return total / len(scores)

def find_highest(scores):
    if not scores: # Handle empty list case
        return None
    highest = scores[0]
    for s in scores:
        if s > highest:
            highest = s
    return highest

def find_lowest(scores):
    if not scores: # Handle empty list case
        return None
    lowest = scores[0]
    for s in scores:
        if s < lowest:
            lowest = s
    return lowest

def process_scores(scores):
    if not scores:
        print("No scores to process.")
        return

    avg = calculate_average(scores)
    highest = find_highest(scores)
    lowest = find_lowest(scores)

    print("Average:", avg)
    print("Highest:", highest)
    print("Lowest:", lowest)
````
