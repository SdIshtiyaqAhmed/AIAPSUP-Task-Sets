**Gemini (AI) Generated Code:**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        """Inserts a new node at the beginning of the list."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        """Inserts a new node at the end of the list."""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def display(self):
        """Prints the elements of the list."""
        elements = []
        current = self.head
        while current:
            elements.append(current.data)
            current = current.next
        print(" -> ".join(map(str, elements)))

# Example Usage:
my_list = SinglyLinkedList()
print("Inserting at end: 10, 20, 30")
my_list.insert_at_end(10)
my_list.insert_at_end(20)
my_list.insert_at_end(30)
print("List after inserting at end:")
my_list.display()

print("\nInserting at beginning: 5")
my_list.insert_at_beginning(5)
print("List after inserting at beginning:")
my_list.display()

print("\nInserting another at beginning: 1")
my_list.insert_at_beginning(1)
print("List after another insert at beginning:")
my_list.display()

print("\nInserting another at end: 40")
my_list.insert_at_end(40)
print("List after another insert at end:")
my_list.display()
```
