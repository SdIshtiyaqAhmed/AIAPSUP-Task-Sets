**Prompt:** write code for the function named is_prime(), that takes a number as an argument and checks whether the number is a prime number or not

**Co-Pilot (AI) Generated Code:**

```python
def is_prime(number):
    # Check if number is less than 2 (not prime)
    if number < 2:
        print(f"{number} is not a prime number")
        return False
    
    # Check for divisibility from 2 to square root of number
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            print(f"{number} is not a prime number")
            return False
    
    print(f"{number} is a prime number")
    return True

# Example usage:
is_prime(29)
is_prime(15)
```
