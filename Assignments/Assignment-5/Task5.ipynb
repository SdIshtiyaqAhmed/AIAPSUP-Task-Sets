**Gemini (AI) Generated Code:**

````python
def greet_user(name, gender=None):
    """Greets the user based on their name and optional gender.

    Args:
        name: The name of the user.
        gender: The gender of the user ('male', 'female'), or None for gender-neutral.

    Returns:
        A greeting string.
    """
    if gender and gender.lower() == "male":
        title = "Mr."
    elif gender and gender.lower() == "female":
        title = "Ms." # Using Ms. as a neutral term for females
    else:
        title = "" # No title for gender-neutral greeting

    if title:
        return f"Hello, {title} {name}! Welcome."
    else:
        return f"Hello, {name}! Welcome."
````
