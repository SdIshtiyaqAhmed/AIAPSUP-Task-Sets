**Prompt:** generate test cases to test the convert_date_format() function which takes a string (date_str) as argument and converts the date_str from "YYYY-MM-DD" to "DD-MM-YYYY"

**Code:**

```python
def convert_date_format(date_str):
  if type(date_str) == type(str()) and len(date_str) == 10 and date_str[4] == "-" and date_str[7] == "-" and len(date_str) == 10:
    temp = date_str.split("-")
    new_date_str = ""
    for i in range(len(temp)-1,-1,-1):
      if i == 0:
        new_date_str += temp[i]
      else:
        new_date_str += temp[i] + "-"
    return new_date_str
  
  else:
    return "Invalid input"
```
