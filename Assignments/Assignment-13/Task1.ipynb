**Gemini (AI) Generated Code:**

````python
import math

def calculate_area_rectangle(length, width):
    return length * width

def calculate_area_square(side):
    return side * side

def calculate_area_circle(radius):
    return math.pi * radius * radius

def calculate_area_refactored(shape, x, y=0):
    area_dispatch = {
        "rectangle": lambda: calculate_area_rectangle(x, y),
        "square": lambda: calculate_area_square(x),
        "circle": lambda: calculate_area_circle(x)
    }

    if shape in area_dispatch:
        return area_dispatch[shape]()
    else:
        raise ValueError(f"Invalid shape: {shape}")

# Example Usage:
print(f"Area of rectangle (5, 10): {calculate_area_refactored('rectangle', 5, 10)}")
print(f"Area of square (7): {calculate_area_refactored('square', 7)}")
print(f"Area of circle (3): {calculate_area_refactored('circle', 3)}")

# Example of invalid shape
try:
    print(calculate_area_refactored('triangle', 5, 4))
except ValueError as e:
    print(e)
````
