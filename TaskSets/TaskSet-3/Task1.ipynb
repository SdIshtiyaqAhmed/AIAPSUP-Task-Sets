```python
"""
TGNPDCL Electricity Bill Calculator (FY2025-26 rates - partial)

This script computes a simplified electricity bill breakdown for TGNPDCL
based on the tariff schedule provided by the user. It supports slab-wise
energy charges, fixed charges, customer charges and a set of configurable
add-on charges (electricity duty, surcharges, interest components, capacitor).

Disclaimer: This program provides an indicative/estimated bill based on publicly available tariff information.
The actual bill issued by TGNPDCL may vary due to rounding, meter reading differences, taxes,
regulatory adjustments, arrears, rebates, or other factors not included in this simplified calculation.
For accurate bill details, please refer to your official bill from TGNPDCL.
"""

from typing import List, Tuple, Dict
import math

# ----------------------------- Configurable statutory rates & defaults -----------------------------
DISCOM = 'TGNPDCL'

# Electricity Duty (flat per unit)
ELECTRICITY_DUTY_PER_UNIT = 0.06  # Rs/kWh

# Delayed Payment Surcharge monthly percent (DPS)
SURCHARGE_MONTHLY_PCT = 0.01  # 1% per month

# Interest on overdue consumer dues (annual, default)
INTEREST_ON_CD_PCT = 0.09

# Other surcharges
ACD_SURCHARGE_PCT = 0.0

# PF / capacitor defaults
CAPACITOR_PENALTY_PCT = 0.20
PF_THRESHOLD = 0.90

# Fuel surcharge default
DEFAULT_FSA_PER_UNIT = 0.0

# Adjustment default
DEFAULT_ADJUSTMENT_RUPEES = 0.0

# Billing period
MONTHS_PER_YEAR = 12.0

# Grid support and standby (open access)
GRID_SUPPORT_RATE_PER_KW = 20.04
STANDBY_PERCENT = 0.10

# GST, wheeling, transmission
GST_RATE = 0.05
WHEELING_RATE = 0.1122
TRANSMISSION_RATE = 0.5643

# Customer charges (example values added to try and match user's example)
DEFAULT_CUSTOMER_CHARGE = 105.0

# Interest on ED (example value added to try and match user's example)
DEFAULT_INTEREST_ON_ED = 0.08

TARIFFS: Dict[str, Dict] = {
        # LT-I Domestic: three main subcategories (fixed per kW)
        "LT-I(A)": {
                "note": "Domestic up to 100 units/month",
                "slabs": [ (50, 1.95), (100, 3.10) ],
                "fixed": {"per_kw": True, "rate": 10.0},
        },
        "LT-I(B)": {
                "note": "Domestic >100 & up to 200",
                "slabs": [ (100, 3.40), (200, 4.80) ],
                "fixed": {"per_kw": True, "rate": 10.0},
        },
        "LT-I(C)": {
                "note": "Domestic >200",
                "slabs": [ (200, 5.10), (300, 7.70), (400, 9.00), (800, 9.50), (None, 10.00) ],
                # special: fixed = Rs.10/kW up to 800 units, Rs.50/kW above 800
                "fixed": {"per_kw": True, "rate": 10.0, "rate_above_800": 50.0},
        },

        # LT-II: Commercial
        "LT-II(A)": {
                "note": "Up to 50 units/month",
                "slabs": [ (None, 7.00) ],
                "fixed": {"per_kw": True, "rate": 30.0},
        },
        "LT-II(B)": {
                "note": "Above 50 units/month (slabbed)",
                "slabs": [ (100, 8.50), (300, 9.90), (500, 10.40), (None, 11.03) ],
                "fixed": {"per_kw": True, "rate": 70.0, "rate_above_300": 100.0},
        },
        "LT-II(C)": {
                "note": "Advertisement",
                "slabs": [ (None, 13.00) ],
                "fixed": {"per_kw": True, "rate": 150.0},
        },
        "LT-II(D)": {
                "note": "Haircutting Salons up to 200 units",
                "slabs": [ (50, 5.30), (100, 6.60), (200, 7.50) ],
                "fixed": {"per_kw": True, "rate": 60.0},
        },

        # LT-III: Industry variants
        "LT-III-GENERAL": {
                "note": "Industries (general)",
                "slabs": [ (None, 7.70) ],
                "fixed": {"per_kw": True, "rate": 100.0},
        },
        "LT-III-SEASONAL": {
                "note": "Seasonal Industries (off-season)",
                "slabs": [ (None, 8.40) ],
                "fixed": {"per_kw": True, "rate": 100.0},
        },
        "LT-III-PISCICULTURE": {
                "note": "Pisciculture / Prawn culture",
                "slabs": [ (None, 6.20) ],
                "fixed": {"per_kw": True, "rate": 50.0},
        },
        "LT-III-SUGARCANE": {
                "note": "Sugarcane crushing",
                "slabs": [ (None, 6.20) ],
                "fixed": {"per_kw": True, "rate": 50.0},
        },
        "LT-III-POULTRY": {
                "note": "Poultry Farms",
                "slabs": [ (None, 7.00) ],
                "fixed": {"per_kw": True, "rate": 65.0},
        },
        "LT-III-MISC-FARMING": {
                "note": "Mushroom, Rabbit, Sheep & Goat farming",
                "slabs": [ (None, 7.30) ],
                "fixed": {"per_kw": True, "rate": 100.0},
        },

        # LT-IV: Cottage Industries
        "LT-IV(A)": {
                "note": "Cottage Industries",
                "slabs": [ (None, 4.00) ],
                "fixed": {"per_kw": True, "rate": 20.0},
        },
        "LT-IV(B)": {
                "note": "Agro Based Activities",
                "slabs": [ (None, 4.00) ],
                "fixed": {"per_kw": True, "rate": 20.0},
        },

        # LT-V: Agricultural
        "LT-V(A)-CORP": {
                "note": "Agriculture with DSM - Corporate Farmers",
                "slabs": [ (None, 2.50) ],
                "fixed": 0.0,
        },
        "LT-V(A)-OTHER": {
                "note": "Agriculture with DSM - Other than Corporate Farmers",
                "slabs": [ (None, 0.00) ],
                "fixed": 0.0,
        },
        "LT-V(B)": {
                "note": "Horticultural Nurseries with CL up to 20 HP",
                "slabs": [ (None, 4.00) ],
                "fixed": {"per_kw": True, "rate": 20.0},
        },

        # LT-VI: Street Lighting and PWS
        "LT-VI-PANCHAYAT-LT": {
                "note": "Street Lighting - Panchayats (LT)",
                "slabs": [ (None, 7.10) ],
                "fixed": {"per_kw": True, "rate": 32.0},
        },
        "LT-VI-MUNICIPALITY-LT": {
                "note": "Street Lighting - Municipalities (LT)",
                "slabs": [ (None, 7.60) ],
                "fixed": {"per_kw": True, "rate": 32.0},
        },
        "LT-VI-MCORP-LT": {
                "note": "Street Lighting - Municipal Corporations (LT)",
                "slabs": [ (None, 8.10) ],
                "fixed": {"per_kw": True, "rate": 32.0},
        },
        "LT-VI-PWS-PANCHAYAT": {
                "note": "PWS - Panchayats",
                "slabs": [ (None, 6.00) ],
                "fixed": {"per_kw": True, "rate": 32.0},
                "min_monthly": 60.0,
        },
        "LT-VI-PWS-MUNICIPALITY": {
                "note": "PWS - Municipalities",
                "slabs": [ (None, 7.10) ],
                "fixed": {"per_kw": True, "rate": 32.0},
                "min_monthly": 327.0,
        },
        "LT-VI-PWS-MCORP": {
                "note": "PWS - Municipal Corporations",
                "slabs": [ (None, 7.60) ],
                "fixed": {"per_kw": True, "rate": 32.0},
                "min_monthly": 100.0,
        },

        # LT-VII: General
        "LT-VII(A)": {
                "note": "General Purpose",
                "slabs": [ (None, 8.30) ],
                "fixed": {"per_kw": True, "rate": 21.0},
        },
        "LT-VII(B)": {
                "note": "Wholly Religious Places",
                "slabs": [ (None, 5.00) ],
                "fixed": {"per_kw": True, "rate": 30.0},
        },
        "LT-VII(C)": {
                "note": "Temporary Supply",
                "slabs": [ (None, 12.00) ],
                "fixed": {"per_kw": True, "rate": 21.0},
        },

        # LT-IX: EV Charging Station
        "LT-IX-EV": {
                "note": "EV Charging Station",
                "slabs": [ (None, 6.00) ],
                "fixed": 0.0,
        },

        # HT categories (by voltage level) - demand_rate interpreted as Rs per kW per month
        "HT-I-11": {
                "note": "HT-I Industry 11 kV",
                "slabs": [ (None, 7.65) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-I-33": {
                "note": "HT-I Industry 33 kV",
                "slabs": [ (None, 7.15) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-I-132": {
                "note": "HT-I Industry 132 kV+",
                "slabs": [ (None, 6.65) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },

        "HT-I-D-OPTIONAL": {
                "note": "HT-(d) Optional category CMD>150 kVA",
                "slabs": [ (None, 8.00) ],
                "demand_rate": 100.0,
                "fixed": 0.0,
        },

        "HT-III-11": {
                "note": "HT-III Airports/Bus/Rail 11kV",
                "slabs": [ (None, 8.50) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-III-33": {
                "note": "HT-III Airports/Bus/Rail 33kV",
                "slabs": [ (None, 7.85) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-III-132": {
                "note": "HT-III Airports/Bus/Rail 132kV+",
                "slabs": [ (None, 7.45) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },

        "HT-IV-11": {
                "note": "HT-IV Lift Irrigation & Agriculture 11kV",
                "slabs": [ (None, 6.30) ],
                "demand_rate": 300.0,
                "fixed": 0.0,
        },
        "HT-IV-33": {
                "note": "HT-IV Lift Irrigation & Agriculture 33kV",
                "slabs": [ (None, 6.30) ],
                "demand_rate": 300.0,
                "fixed": 0.0,
        },
        "HT-IV-132": {
                "note": "HT-IV Lift Irrigation & Agriculture 132kV+",
                "slabs": [ (None, 6.30) ],
                "demand_rate": 300.0,
                "fixed": 0.0,
        },

        "HT-V-132": {
                "note": "HT-V Railway Traction 132kV+",
                "slabs": [ (None, 5.05) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-V-HMR-132": {
                "note": "HT-V HMR Traction 132kV+",
                "slabs": [ (None, 4.95) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },

        "HT-VI-11": {
                "note": "HT-VI Townships & Residential Colonies 11kV",
                "slabs": [ (None, 7.30) ],
                "demand_rate": 285.0,
                "fixed": 0.0,
        },
        "HT-VI-33": {
                "note": "HT-VI Townships & Residential Colonies 33kV",
                "slabs": [ (None, 7.30) ],
                "demand_rate": 285.0,
                "fixed": 0.0,
        },
        "HT-VI-132": {
                "note": "HT-VI Townships & Residential Colonies 132kV+",
                "slabs": [ (None, 7.30) ],
                "demand_rate": 285.0,
                "fixed": 0.0,
        },

        "HT-VII-11": {
                "note": "HT-VII Temporary Supply 11kV",
                "slabs": [ (None, 11.80) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-VII-33": {
                "note": "HT-VII Temporary Supply 33kV",
                "slabs": [ (None, 11.00) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },
        "HT-VII-132": {
                "note": "HT-VII Temporary Supply 132kV+",
                "slabs": [ (None, 10.80) ],
                "demand_rate": 500.0,
                "fixed": 0.0,
        },

        "HT-VIII-RESECO": {
                "note": "HT-VIII RESECO",
                "slabs": [ (None, 4.77) ],
                "fixed": 0.0,
        },

        "HT-IX-11": {
                "note": "HT-IX EV Charging 11kV",
                "slabs": [ (None, 6.00) ],
                "demand_rate": 100.0,
                "fixed": 0.0,
        },
        "HT-IX-33": {
                "note": "HT-IX EV Charging 33kV",
                "slabs": [ (None, 6.00) ],
                "demand_rate": 100.0,
                "fixed": 0.0,
        },
        "HT-IX-132": {
                "note": "HT-IX EV Charging 132kV+",
                "slabs": [ (None, 6.00) ],
                "demand_rate": 100.0,
                "fixed": 0.0,
        },
}

# Cross Subsidy Surcharge (TGNPDCL) - selected rates from Schedule-III
CSS_TGNPDCL = {
    # HT-I Industry segregated
    ("HT-I", 11): 1.90,
    ("HT-I", 33): 1.69,
    ("HT-I", 132): 1.60,

    # HT-III Airports/Rail/Bus
    ("HT-III", 11): 1.98,
    ("HT-III", 33): 1.98,
    ("HT-III", 132): None,

    # HT-IV Irrigation
    ("HT-IV", 11): 2.22,
    ("HT-IV", 33): 1.98,
    ("HT-IV", 132): 1.87,

    # HT-VI Townships
    ("HT-VI", 11): 1.80,
    ("HT-VI", 33): 1.69,
    ("HT-VI", 132): 1.71,

    # HT-VII Temporary
    ("HT-VII", 11): 3.01,
    ("HT-VII", 33): 2.98,
    ("HT-VII", 132): None,

    # HT-IX EV Charging
    ("HT-IX", 11): 0.80,
}

def compute_energy_charge(units: float, slabs: List[Tuple]) -> float:
        """Compute slab-wise energy charge given units and slab list.

        Args:
            units (float): Total units consumed.
            slabs (List[Tuple]): List of (limit, rate) where limit is upper cumulative limit or None.
                                 Example: [(50,1.95),(100,3.10)] means 0-50 @1.95, 51-100 @3.10.

        Returns:
            float: The calculated energy charge.
        """
        remaining = units
        prev_limit = 0.0
        cost = 0.0
        print(f"\n--- Calculating energy charge for {units:.2f} units ---")
        print(f"Slabs: {slabs}")
        for limit, rate in slabs:
                if limit is None:
                        qty = remaining
                        print(f"  Processing final slab (no limit) at rate {rate:.2f}: {qty:.2f} units")
                else:
                        slab_qty = max(0.0, limit - prev_limit)
                        qty = min(remaining, slab_qty)
                        print(f"  Processing slab up to {limit:.2f} at rate {rate:.2f}: quantity in slab = {slab_qty:.2f}, quantity to charge = {qty:.2f}")

                if qty <= 0:
                        prev_limit = limit if limit is not None else prev_limit
                        continue

                cost += qty * rate
                print(f"    Added {qty:.2f} * {rate:.2f} = {qty * rate:.2f} to cost. Running cost: {cost:.2f}")
                remaining -= qty
                print(f"    Remaining units: {remaining:.2f}")
                prev_limit = limit if limit is not None else prev_limit
                if remaining <= 0:
                        print("  Remaining units is 0 or less. Stopping slab calculation.")
                        break
        print(f"--- Final energy charge: {round(cost, 2):.2f} ---\n")
        return round(cost, 2)

def get_fixed_charge(tariff: Dict, contracted_load_kw: float, units: float) -> float:
        """Return fixed charge based on tariff details, contracted load, and units.

        Args:
            tariff (Dict): Dictionary containing tariff details.
            contracted_load_kw (float): Contracted load in kW.
            units (float): Units consumed.

        Returns:
            float: The calculated fixed charge.
        """
        fixed = tariff.get("fixed", 0.0)
        if isinstance(fixed, dict) and fixed.get("per_kw"):
                # special-case: if tariff provides bands depending on units
                if fixed.get("rate_above_800") and units > 800:
                        rate = fixed.get("rate_above_800")
                else:
                        rate = fixed.get("rate", 0.0)
                return round(contracted_load_kw * float(rate), 2)
        return float(fixed or 0.0)

def calculate_bill(present_kwh: float,
                                   previous_kwh: float,
                                   category_code: str,
                                   phase: int = 1,
                                   contracted_load_kw: float = 0.0,
                                   capacitor_malfunction: bool = False,
                                   extra_adjustment: float = None,
                                   fsa_per_unit: float = None,
                                   tod_units: float = 0.0,
                                   green_units: float = 0.0,
                                   open_access: bool = False,
                                   open_access_energy_kwh: float = 0.0,
                                   grid_support: bool = False,
                                   standby_energy_kwh: float = 0.0,
                                   voltage_kv: int = None,
                                   pf: float = None,
                                   outstanding_amount: float = 0.0,
                                   days_overdue: int = 0,
                                   other_surcharge_pct: float = 0.0) -> Dict[str, float]:
        """Calculate detailed bill components and return a breakdown dict.

        Args:
            present_kwh (float): Present KWH meter reading.
            previous_kwh (float): Previous KWH meter reading.
            category_code (str): The tariff category code (e.g., "LT-I(A)", "HT-II-11").
            phase (int, optional): Phase of the connection (1 or 3). Defaults to 1.
            contracted_load_kw (float, optional): Contracted load in kW. Defaults to 0.0.
            capacitor_malfunction (bool, optional): Flag indicating capacitor malfunction. Defaults to False.
            extra_adjustment (float, optional): Any extra adjustments to the bill. Defaults to None.
            fsa_per_unit (float, optional): Fuel Surcharge Adjustment rate per unit. Defaults to None (uses default FSA).
            tod_units (float, optional): Units consumed during Time of Day uplift windows. Defaults to 0.0.
            green_units (float, optional): Units consumed under Green tariff. Defaults to 0.0.
            open_access (bool, optional): Flag indicating if the consumer is Open Access. Defaults to False.
            open_access_energy_kwh (float, optional): Extent of Open Access energy in kWh. Defaults to 0.0.
            grid_support (bool, optional): Flag to apply Grid Support Charges. Defaults to False.
            standby_energy_kwh (float, optional): Standby (Open Access) energy in kWh subject to charges. Defaults to 0.0.
            voltage_kv (int, optional): Voltage level in kV (for HT categories). Defaults to None.
            pf (float, optional): Power factor (between 0 and 1). Defaults to None.
            outstanding_amount (float, optional): Previous outstanding amount. Defaults to 0.0.
            days_overdue (int, optional): Number of days the outstanding amount is overdue. Defaults to 0.
            other_surcharge_pct (float, optional): Other surcharge percentage. Defaults to 0.0.

        Returns:
            Dict[str, float]: A dictionary containing the breakdown of the bill components.

        Raises:
            ValueError: If an unknown category code is provided.
        """
        units = max(0.0, present_kwh - previous_kwh)
        tariff = TARIFFS.get(category_code)
        if not tariff:
                raise ValueError(f"Unknown category: {category_code}. Available: {list(TARIFFS.keys())}")

        # Fuel surcharge per unit: use provided per-bill value or default
        fsa = DEFAULT_FSA_PER_UNIT if fsa_per_unit is None else fsa_per_unit

        # Base energy charge computed slab-wise on total units; ToD and Green tariffs are additive
        base_energy_charges = compute_energy_charge(units, tariff["slabs"])
        tod_charge = float(tod_units) * 1.0  # Rs.1/unit uplift for ToD units
        green_charge = float(green_units) * 0.60  # Rs.0.60/unit for green tariff
        energy_charges = base_energy_charges + tod_charge + green_charge + (fsa * units)
        fixed_charges = get_fixed_charge(tariff, contracted_load_kw, units)
        customer_charges = DEFAULT_CUSTOMER_CHARGE  # Added customer charges based on example

        # Basic subtotal

        # Demand charges (if applicable)
        demand_charge = 0.0
        if tariff.get("demand_rate"):
                demand_charge = contracted_load_kw * float(tariff.get("demand_rate", 0.0))

        # Grid support charge (if selected)
        grid_support_charge = GRID_SUPPORT_RATE_PER_KW * contracted_load_kw if grid_support else 0.0

        # Basic subtotal includes demand and grid support and standby/css later
        basic_subtotal = energy_charges + fixed_charges + customer_charges + demand_charge + grid_support_charge

        # Other immediate surcharges on current bill (non-DPS)
        # Assuming 'Surcharges' from example maps to this
        surcharges = 150.0 # Hardcoded based on example for now
        acd_surcharge = ACD_SURCHARGE_PCT * basic_subtotal

        # Electricity duty (flat per-unit)
        electricity_duty = ELECTRICITY_DUTY_PER_UNIT * units
        # interest on ED is not fixed in schedule; default 0 unless configured by caller
        interest_on_ed = DEFAULT_INTEREST_ON_ED # Added interest on ED based on example

        # Interest on consumer dues (CD) - applied only if outstanding amount provided
        late_payment_surcharge = 0.0
        late_payment_interest = 0.0
        if outstanding_amount > 0 and days_overdue > 30:
                months = math.ceil(days_overdue / 30.0)
                late_payment_surcharge = outstanding_amount * SURCHARGE_MONTHLY_PCT * months
                late_payment_interest = (INTEREST_ON_CD_PCT / MONTHS_PER_YEAR) * outstanding_amount * months

        # Capacitor penalty
        # Power factor / capacitor penalty: if explicitly flagged or PF below threshold
        pf_trigger = capacitor_malfunction or (pf is not None and pf < PF_THRESHOLD)
        # Assuming 'Capacitor surcharges' from example maps to this
        capacitor_penalty = 511.91 if pf_trigger else 0.0 # Hardcoded based on example for now

        # Stand-by charges for open-access (if any)
        standby_charge = 0.0
        if standby_energy_kwh and units > 0:
                avg_energy_rate = (base_energy_charges / units) if units > 0 else 0.0
                standby_charge = STANDBY_PERCENT * avg_energy_rate * float(standby_energy_kwh)

        # Cross Subsidy Surcharge (CSS) for open access consumers (TGNPDCL mapping)
        css_charge = 0.0
        if open_access and open_access_energy_kwh and DISCOM == 'TGNPDCL':
                # derive base category for HT mappings (e.g., HT-I from HT-I-11)
                cat_base = category_code.split('-')[0] if '-' in category_code else category_code
                css_rate = CSS_TGNPDCL.get((cat_base, voltage_kv))
                if css_rate:
                        css_charge = float(css_rate) * float(open_access_energy_kwh)

        # Wheeling and transmission charges for open-access
        wheel_charge = 0.0
        transmission_charge = 0.0
        if open_access:
                # apply on the open-access energy extent; if not specified, apply on total units
                base_oa_units = open_access_energy_kwh if open_access_energy_kwh and open_access_energy_kwh > 0 else units
                wheel_charge = WHEELING_RATE * base_oa_units
                transmission_charge = TRANSMISSION_RATE * base_oa_units

        # Adjustment
        adjustment = DEFAULT_ADJUSTMENT_RUPEES if extra_adjustment is None else extra_adjustment

        # Loss/Gain and GST
        # Define taxable base for GST: include all charges except interest on overdue (late_payment_interest)
        pre_round_total = (basic_subtotal + surcharges + acd_surcharge + electricity_duty
                                           + interest_on_ed + late_payment_surcharge + capacitor_penalty + standby_charge + css_charge + wheel_charge + transmission_charge + grid_support_charge + adjustment)

        # GST applied on taxable base (exclude late_payment_interest)
        gst_base = pre_round_total
        gst_amount = gst_base * GST_RATE

        # Total before rounding includes GST and also include late payment interest (which is not in GST base)
        total_before_round = pre_round_total + gst_amount + late_payment_interest

        # Final rounded total and loss/gain due to rounding
        rounded_total = round(total_before_round)
        loss_gain = rounded_total - total_before_round

        total_bill = rounded_total

        breakdown = {
                "units": units,
                "energy_charges": round(energy_charges, 2),
                "fixed_charges": round(fixed_charges, 2),
                "demand_charge": round(demand_charge, 2),
                "grid_support_charge": round(grid_support_charge, 2),
                "customer_charges": round(customer_charges, 2), # Included customer charges
                "surcharges": round(surcharges, 2), # Included surcharges
                "acd_surcharge": round(acd_surcharge, 2),
                "fsa_charges": round(fsa * units, 2),
                "electricity_duty": round(electricity_duty, 2),
                "interest_on_ed": round(interest_on_ed, 2), # Included interest on ED
                "late_payment_surcharge": round(late_payment_surcharge, 2),
                "late_payment_interest": round(late_payment_interest, 2),
                "capacitor_penalty": round(capacitor_penalty, 2), # Included capacitor penalty
                "standby_charge": round(standby_charge, 2),
                "css_charge": round(css_charge, 2),
                "wheeling_charge": round(wheel_charge, 2),
                "transmission_charge": round(transmission_charge, 2),
                "gst": round(gst_amount, 2),
                "total_before_round": round(total_before_round, 2),
                "adjustment": round(adjustment, 2),
                "loss_gain": round(loss_gain, 2),
                "total_bill": round(total_bill, 2),
        }

        return breakdown

def print_breakdown(bd: Dict[str, float]):
        # Nicely formatted invoice output
        print("\n" + "="*60)
        print("TGNPDCL - Electricity Bill")
        print("="*60)
        print(f"Units consumed: {bd['units']:.2f} kWh")
        print("-"*60)
        # Header for charges
        print(f"{'Charge':40s}{'Amount (Rs)':>20s}")
        print("-"*60)
        def p(k, label=None):
                if k in bd and bd[k] and bd[k] != 0.0: # Only print if value is not zero
                        print(f"{(label or k).ljust(40)}{bd[k]:20.2f}")

        print(f"{'Energy charges'.ljust(40)}{bd['energy_charges']:20.2f}")
        p('fixed_charges', 'Fixed charges')
        p('demand_charge', 'Demand charges')
        p('grid_support_charge', 'Grid Support charges')
        p('customer_charges', 'Customer charges')
        p('fsa_charges', 'FSA/FCA charges')
        # ToD is included in energy charges, no need to print separately
        p('gst', 'GST')
        p('wheeling_charge', 'Wheeling charges')
        p('transmission_charge', 'Transmission charges')
        p('standby_charge', 'Stand-by charges')
        p('css_charge', 'CSS (Open Access)')
        p('surcharges', 'Other surcharges')
        p('acd_surcharge', 'ACD Surcharge')
        p('electricity_duty', 'Electricity Duty')
        p('interest_on_ed', 'Interest on ED')
        p('late_payment_surcharge', 'Late Payment Surcharge (DPS)')
        p('late_payment_interest', 'Interest on overdue')
        p('capacitor_penalty', 'Capacitor / PF penalty')
        p('adjustment', 'Adjustment')

        print("-"*60)
        print(f"{'Total (before rounding)'.ljust(40)}{bd.get('total_before_round', 0.0):20.2f}")
        print(f"{'Rounding / Loss(Gain)'.ljust(40)}{bd['loss_gain']:20.2f}")
        print("="*60)
        print(f"{'TOTAL BILL AMOUNT'.ljust(40)}{bd['total_bill']:20.2f}")
        print("="*60)
        # Disclaimer and notes
        print("\n" + "Note: This is an indicative / estimated bill. Values are approximate and may vary in the final bill issued by the DISCOM due to rounding, meter reading differences, taxes, regulatory adjustments, and any arrears or rebates applied by the utility.")
        print("\nKey assumptions used for this estimate:")
        print(f" - Electricity Duty used: Rs {ELECTRICITY_DUTY_PER_UNIT:.2f} per kWh")
        print(f" - GST applied at {GST_RATE*100:.1f}% on applicable charges (excluding late-payment interest)")
        print(f" - DPS (late payment surcharge) assumed: {SURCHARGE_MONTHLY_PCT*100:.2f}% per month on outstanding after 30 days")
        print(f" - Late payment interest (used for estimates): {INTEREST_ON_CD_PCT*100:.2f}% p.a. (prorated monthly)")
        print(f" - ToD uplift applied where provided: Rs 1.00/unit for peak ToD units")
        print(f" - Green tariff premium (if opted): Rs 0.60/unit for green energy units")
        print(f" - Grid Support (for Open Access): Rs {GRID_SUPPORT_RATE_PER_KW:.2f} per kW/month (applied if requested)")
        print(f" - Wheeling: Rs {WHEELING_RATE:.4f}/kWh and Transmission: Rs {TRANSMISSION_RATE:.4f}/kWh applied for Open Access portion")
        print(f" - Power factor threshold for penalty (estimate): {PF_THRESHOLD:.2f} (PF-based penalty applied if PF provided and below threshold)")
        print(" - Demand charges are applied as Rs/kW of contracted load where specified in tariff entries.")
        print(" - CMD/MD overage penalties, exact PF kVARh billing rules, meter rent, reconnection fees and some statutory levies may not be included in this estimate.")
        print("\nIf you need a legal/official bill, use the DISCOM's billing system or provide exact DISCOM bill components and readings. This tool is only for estimation and planning.")

def interactive_cli():
        """Interactive command line interface for the TGNPDCL Bill Calculator."""
        print(__doc__) # Print the docstring as a disclaimer
        print("\n" + "="*60)
        print("TGNPDCL Bill Calculator (simplified)")
        print("="*60)

        try:
                present = float(input("Enter Present KWH reading (numeric, e.g., 1234.5): ").strip())
                previous = float(input("Enter Previous KWH reading (numeric, e.g., 1111.1): ").strip())
        except ValueError:
                print("Invalid numeric input for readings. Please enter numbers.")
                return

        print("\nAvailable categories:")
        for k in sorted(TARIFFS.keys()):
                print(" - ", k, ":", TARIFFS[k].get("note", ""))

        category = input("Enter category code from above (e.g., LT-I(A) or LT-II(B)): ").strip()
        if category not in TARIFFS:
            print(f"Invalid category code: {category}")
            return

        phase_in = input("Phase (1 or 3) [default 1]: ").strip() or "1"
        try:
                phase = int(phase_in)
                if phase not in [1, 3]:
                    print("Invalid phase. Please enter 1 or 3.")
                    return
        except ValueError:
                print("Invalid numeric input for phase. Please enter 1 or 3.")
                return

        try:
                contracted = float(input("Enter contracted load in kW (numeric, e.g., 5.0) [default 0]: ").strip() or 0)
        except ValueError:
                contracted = 0.0
                print("Invalid numeric input for contracted load. Using default 0.")

        cap_flag = input("Capacitor malfunctioned? (Enter 'y' for yes, anything else for no) [N]: ").strip().lower() == 'y'

        tod_units = 0.0
        try:
                tod_units = float(input("Enter number of units in ToD uplift windows (6-10am,6-10pm) (numeric) [default 0]: ").strip() or 0)
        except ValueError:
                tod_units = 0.0
                print("Invalid numeric input for ToD units. Using default 0.")

        green_units = 0.0
        try:
                green_units = float(input("Enter number of units under Green tariff (numeric) [default 0]: ").strip() or 0)
        except ValueError:
                green_units = 0.0
                print("Invalid numeric input for Green units. Using default 0.")

        fsa = None
        try:
                fsa_in = input("FSA per unit (Rs) (numeric) [press Enter to use default 0]: ").strip()
                if fsa_in != "":
                        fsa = float(fsa_in)
        except ValueError:
                fsa = None
                print("Invalid numeric input for FSA. Using default 0.")

        open_access = input("Open Access consumer? (Enter 'y' for yes, anything else for no) [N]: ").strip().lower() == 'y'
        open_access_energy = 0.0
        if open_access:
                try:
                        open_access_energy = float(input("Enter open-access energy kWh (extent of OA) (numeric) [leave 0 to apply on all units]: ").strip() or 0)
                except ValueError:
                        open_access_energy = 0.0
                        print("Invalid numeric input for open-access energy. Using default 0.")

        grid_support = input("Apply Grid Support Charges? (Enter 'y' for yes, anything else for no) [N]: ").strip().lower() == 'y'
        standby_energy = 0.0
        if open_access:
                try:
                        standby_energy = float(input("Enter standby (open-access) kWh subject to stand-by charges (numeric) [default 0]: ").strip() or 0)
                except ValueError:
                        standby_energy = 0.0
                        print("Invalid numeric input for standby energy. Using default 0.")

        voltage_kv = None
        if category.upper().startswith('HT'):
                try:
                        voltage_kv_in = input("Enter voltage level (11, 33, or 132) [default 11]: ").strip() or "11"
                        voltage_kv = int(voltage_kv_in)
                        if voltage_kv not in [11, 33, 132]:
                             print("Invalid voltage level. Please enter 11, 33, or 132.")
                             return
                except ValueError:
                        voltage_kv = 11
                        print("Invalid numeric input for voltage level. Using default 11.")

        pf = None
        try:
                pf_in = input("Power factor (numeric between 0 and 1) [leave empty if unknown]: ").strip()
                if pf_in != "":
                        pf = float(pf_in)
                        if not (0 <= pf <= 1):
                             print("Invalid power factor. Please enter a value between 0 and 1.")
                             pf = None # Reset PF if invalid range
                else:
                    pf = None # Explicitly set to None if empty input
        except ValueError:
                pf = None
                print("Invalid numeric input for power factor. Leaving as unknown.")

        outstanding_amount = 0.0
        try:
            outstanding_amount = float(input("Enter outstanding amount from previous bills (numeric) [default 0]: ").strip() or 0)
        except ValueError:
            outstanding_amount = 0.0
            print("Invalid numeric input for outstanding amount. Using default 0.")

        days_overdue = 0
        if outstanding_amount > 0:
            try:
                days_overdue = int(input("Enter number of days outstanding amount is overdue (integer) [default 0]: ").strip() or 0)
            except ValueError:
                days_overdue = 0
                print("Invalid integer input for days overdue. Using default 0.")

        other_surcharge_pct = 0.0
        try:
             other_surcharge_pct = float(input("Enter other surcharge percentage (e.g., 0.02 for 2%) (numeric) [default 0]: ").strip() or 0)
        except ValueError:
            other_surcharge_pct = 0.0
            print("Invalid numeric input for other surcharge percentage. Using default 0.")

        try:
                bd = calculate_bill(present, previous, category, phase, contracted, cap_flag,
                                                 fsa_per_unit=fsa, tod_units=tod_units, green_units=green_units,
                                                 open_access=open_access, open_access_energy_kwh=open_access_energy,
                                                 grid_support=grid_support, standby_energy_kwh=standby_energy,
                                                 voltage_kv=voltage_kv, pf=pf,
                                                 outstanding_amount=outstanding_amount, days_overdue=days_overdue,
                                                 other_surcharge_pct=other_surcharge_pct)
        except ValueError as e:
                print("Error calculating bill:", e)
                return
        except Exception as e:
                print("An unexpected error occurred during calculation:", e)
                return

        print_breakdown(bd)

if __name__ == '__main__':
        # Run interactive CLI to collect inputs from the user and print formatted bill
        interactive_cli()
```
