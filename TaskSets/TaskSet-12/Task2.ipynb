**Gemini (AI) Generated Code:**

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Example usage:
my_list = [64, 34, 25, 12, 22, 11, 90]
print("Original list:", my_list)

sorted_list = bubble_sort(list(my_list)) # Pass a copy to keep original list intact
print("Sorted list:", sorted_list)

# Another example
my_list_2 = [5, 1, 4, 2, 8]
print("Original list 2:", my_list_2)

sorted_list_2 = bubble_sort(list(my_list_2))
print("Sorted list 2:", sorted_list_2)
```
