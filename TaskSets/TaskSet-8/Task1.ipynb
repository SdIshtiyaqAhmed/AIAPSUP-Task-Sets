**Description:** Use AI to generate test cases for is_valid_email(email) and then implement the validator function.  

Requirements:

* Must contain @ and . characters.

* Must not start or end with special characters.

* Should not allow multiple @.

**Expected Output:** Email validation logic passing all test cases


**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Prompt:** Generate test cases to test the function, the requirements for a valid email are, must contain @ and . characters, must not start or end with special characters, should not allow multiple @

**Code:**

```python
def is_valid_email(email):
  letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
  if (type(email) == type(str())) and ('@' and '.' in email) and (email.count('@') == 1) and (email[0] in letters) and (email[-1] in letters) and (email.index('@') != email.index('.')-1 or email.index('@')+1 != email.index('.')):
      return True
  return False
```
