**Prompt:** A is_sentence_palindrome() function that takes a string (sentence) as an argument to check whether the sentence is a palindrome or not. It Ignores the case, punctuation, and spaces when checking for a palindrome. Generate test cases to test the function

**Code:**

```python
def is_sentence_palindrome(sentence):
  if (type(sentence) == type(str())):
    sentence = sentence.lower()
    temp = ""
    letters = "abcdefghijklmnopqrstuvwxyz"
    for letter in sentence:
      if letter in letters:
        temp += letter
    if temp == temp[::-1]:
      return True
    else:
      return False
  return "Invalid Input"
```

**Gemini (AI) Generated Code:**

```python
test_cases = [
    # Palindromes
    ("madam", True),
    ("Madam", True),
    ("A man, a plan, a canal: Panama", True),
    ("Racecar", True),
    ("No lemon, no melon", True),
    ("Was it a car or a cat I saw?", True),
    ("", True), # Empty string can be considered a palindrome

    # Non-palindromes
    ("hello", False),
    ("world", False),
    ("not a palindrome", False),

    # Invalid inputs
    (123, "Invalid Input"),
    (None, "Invalid Input"),
    (['a', 'b', 'a'], "Invalid Input")
]

print("Running test cases for is_sentence_palindrome function:")
for sentence, expected_result in test_cases:
    actual_result = is_sentence_palindrome(sentence)
    if actual_result == expected_result:
        status = "PASSED"
    else:
        status = f"FAILED - Expected: '{expected_result}', Got: '{actual_result}'"
    print(f"Sentence: '{sentence}', Expected: '{expected_result}', Actual: '{actual_result}' - {status}")
```
