**Description:** Ask AI to generate test cases for assign_grade(score) function. Handle boundary and invalid inputs.

Requirements: 

AI should generate test cases for assign_grade(score) where: 90-100: A, 80-89: B, 70-79: C, 60-69: D, <60: F

Include boundary values and invalid inputs (e.g., -5, 105, "eighty").

**Expected Output:** Grade assignment function passing test suite

**Prerequisites:** GitHub Copilot in VS Code / Gemini in Google Colab, Internet Connection

**Prompt:** Generate test cases to test the function for boundary values, and invalid inputs (such as negative score, score more than 100, score is not an integer)

**Code:**

```python
def assign_grade(score):
  if ((type(score) == type(int()))) and (0 <= score <= 100):
    if 90 <= score <= 100:
      return 'A'
    elif 80 <= score <= 89:
      return 'B'
    elif 70 <= score <= 79:
      return 'C'
    elif 60 <= score <= 69:
      return 'D'
    else:
      return 'F'
  return "Invalid Score"
```

**Gemini (AI) Generated Code:**

```python
test_cases = [
    # Boundary values for grades
    (95, 'A'), (90, 'A'), (100, 'A'), # A grade
    (89, 'B'), (80, 'B'), # B grade
    (79, 'C'), (70, 'C'), # C grade
    (69, 'D'), (60, 'D'), # D grade
    (59, 'F'), (50, 'F'), (0, 'F'), # F grade

    # Invalid inputs
    (-5, 'Invalid Score'), # Negative score
    (101, 'Invalid Score'), # Score greater than 100
    ('abc', 'Invalid Score'), # String input
    (None, 'Invalid Score'), # None input
    (30.5, 'Invalid Score'), # Float input (should be invalid as per current function logic)
    (89.9, 'Invalid Score')  # Float input (should be invalid as per current function logic)
]

print("Running test cases for assign_grade function:")
for score, expected_grade in test_cases:
    actual_grade = assign_grade(score)
    if actual_grade == expected_grade:
        status = "PASSED"
    else:
        status = "FAILED"
    print(f"Score: {score}, Expected: '{expected_grade}', Actual: '{actual_grade}' - {status}")
```
