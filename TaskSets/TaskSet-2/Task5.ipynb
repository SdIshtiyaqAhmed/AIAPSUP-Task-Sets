**Code (Program.py):**

```python
def sum_of_even_numbers_and_odd_numbers_in_list(list):
    even_numbers_sum = 0
    odd_numbers_sum = 0
    n = len(list)

    for i in range(n):
        if list[i] % 2 == 0:
            even_numbers_sum += list[i]
        else:
            odd_numbers_sum += list[i]
    
    return even_numbers_sum, odd_numbers_sum

list1 = [1,2,3,5,8,9,13,45]
even_numbers_sum, odd_numbers_sum = sum_of_even_numbers_and_odd_numbers_in_list(list1)
print("Sum of even numbers in the list:", even_numbers_sum)
print("Sum of odd numbers in the list:", odd_numbers_sum)
```
