**Prompt:** write a function to print first 10 multiples of a number using a loop

**Gemini (AI) Generated (Auto-completed) Code:**

```python
def print_multiples(number):
  """Prints the first 10 multiples of a given number.

  Args:
    number: The number for which to print the multiples.
  """
  print(f"Multiples of {number}:")
  for i in range(1, 11):
    print(number * i)

# Example usage:
print_multiples(5)
```

**Prompt:** write a function to print first 10 multiples of a number using other controlled looping

**Gemini (AI) Generated Code:**

```python
def print_multiples(number):
  """Prints the first 10 multiples of a given number using a while loop.

  Args:
    number: The number for which to print the multiples.
  """
  print(f"Multiples of {number}:")
  i = 1
  while i <= 10:
    print(number * i)
    i += 1

# Example usage:
print_multiples(5)
```
