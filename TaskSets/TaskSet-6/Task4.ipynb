```python
def sum_to_n(n):
  """Calculates the sum of the first n natural numbers.

  Args:
    n: An integer representing the upper limit of the sum.

  Returns:
    The sum of the first n natural numbers.
  """
  return n * (n + 1) // 2

# Example usage:
result = sum_to_n(10)
print(f"The sum of the first 10 numbers is: {result}")

result = sum_to_n(100)
print(f"The sum of the first 100 numbers is: {result}")
```

**Gemini (AI) Generated Code:**

```python
def sum_to_n(n):
  """Calculates the sum of the first n natural numbers using a loop.

  Args:
    n: An integer representing the upper limit of the sum.

  Returns:
    The sum of the first n natural numbers.
  """
  total = 0
  for i in range(n + 1):
    total += i
  return total

# Example usage:
result = sum_to_n(10)
print(f"The sum of the first 10 numbers is: {result}")

result = sum_to_n(100)
print(f"The sum of the first 100 numbers is: {result}")
```
