**Gemini (AI) Generated Code:**

````python
class Student:
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores # scores should be a list, e.g., [85, 90, 78]

    def details(self):
        print(f"Name: {self.name}, Age: {self.age}")

    def total_score(self):
        return sum(self.scores)

    def average_score(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)

    def __repr__(self):
        return f"Student(name='{self.name}', age={self.age}, scores={self.scores})"

# Example Usage:
student1 = Student("Alice", 16, [85, 90, 78])
student1.details()
print(f"Total Score: {student1.total_score()}")
print(f"Average Score: {student1.average_score()}")
print(student1)

student2 = Student("Bob", 17, [70, 65, 80, 95]) # Example with more scores
student2.details()
print(f"Total Score: {student2.total_score()}")
print(f"Average Score: {student2.average_score()}")
print(student2)
````
