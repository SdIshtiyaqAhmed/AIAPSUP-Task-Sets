**Description:** Generate both recursive and iterative versions of a factorial function using comments.

**Expected Output:** Two working factorial implementations

**Prerequisites:** Latest version of VS Code is installed, GitHub Copilot in VS Code, Internet Connection

**Prompt:** Iterative and recursive function for factorial

**Co-Pilot (AI) Generated Code:**

```python
# Iterative and recursive function for factorial
def factorial_iterative(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Example usage
if __name__ == "__main__":
    number = 5
    print(f"Iterative Factorial of {number}: {factorial_iterative(number)}")
    print(f"Recursive Factorial of {number}: {factorial_recursive(number)}")
```
