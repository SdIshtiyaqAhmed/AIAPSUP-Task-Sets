**Unsecure Version**

**Prompt:** write code for a simple login system, with username and password as credentials

**Gemini (AI) Generated Code**

**Revised Secure Version**

**Prompt:** write code for a simple login system, with username and password as credentials, and do not hardcode the credentials. Use proper encoding and security for credentials

**Gemini (AI) Generated Code**

```python
# If not installed bcrypt module install it using:
# pip install bcrypt

import bcrypt

users = {} # Initialize or clear the users dictionary for registration

def register_user(username, password):
  """Registers a new user by hashing the password and storing it.

  Args:
    username: The username of the new user.
    password: The password of the new user.

  Returns:
    True if registration is successful, False otherwise.
  """
  if username in users:
    print(f"Username '{username}' already exists.")
    return False

  # bcrypt requires bytes
  password_bytes = password.encode('utf-8')
  hashed_password = bcrypt.hashpw(password_bytes, bcrypt.gensalt())

  users[username] = hashed_password
  print(f"User '{username}' registered successfully.")
  return True
def login(username, password):
  """Checks if the provided username and password match the stored credentials.

  Args:
    username: The username to check.
    password: The password to check.

  Returns:
    True if the username and password match, False otherwise.
  """
  if username in users:
    hashed_password = users[username]
    # bcrypt requires bytes for comparison
    password_bytes = password.encode('utf-8')
    if bcrypt.checkpw(password_bytes, hashed_password):
      return True
    else:
      return False
  else:
    return False
reg_username = input("Enter a username for registration: ")
reg_password = input("Enter a password for registration: ")
register_user(reg_username, reg_password)
login_username = input("Enter your username: ")
login_password = input("Enter your password: ")
is_logged_in = login(login_username, login_password)
if is_logged_in:
  print("Login successful!")
else:
  print("Login failed. Invalid username or password.")
```

**Observations:** The username and passwords are now not hardcoded in the form of a dictionary, but asked as input and then the password is encoded using bcrypt python module. This provides better security to the credentials.
